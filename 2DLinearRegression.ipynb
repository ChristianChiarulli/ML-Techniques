## 2D Linear Regression

Essentially we have two features instead of one influencing our system

A good explaination of multiple linear regression https://newonlinecourses.science.psu.edu/stat501/node/382/

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
```

```python
# go through every line and append the values to the feature vector and true vector

for line in open('data/data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([float(x1), float(x2), 1])      # feature vector
    Y.append(float(y))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
# example of one feature vector remember the 1 is the bias
X[0]
```

```python
# Shape of feature vector 100 rows, 3 columns where each row is a sample and each column is a feature
X.shape
```

```python
# Shape of Y these are the true values
Y.shape
```

```python
fig = plt.figure()

# 111 1 row 1 col 1 pic
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
plt.show()
```

The weight vector can be found using least squares:

$$w = (X'X)^{-1}X'Y$$

where:

* $X'$ is the transpose of the X matrix and,
* $(X'X)^{-1}$ is the inverse of the $X'X$ matrix

https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.solve.html

```python
# calculate weights not using gradient descent
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
```

```python
w.shape
```

```python
# w is a point in three dimensional space
w
```

```python
Yhat = np.dot(X, w)
```

```python
SSres = sum((Y-Yhat)**2)
SStot = sum((Y-Y.mean())**2)
R2 = 1 - SSres/SStot
print('R-squared is: ', R2)
```
